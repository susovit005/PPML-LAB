```python
#convert series from string to numeric  with the error handling
import pandas as pd
s=pd.Series(['10', '20', 'abc', '30'])

# covert to numeric,invalid parsing will be set as NaN
numeric_s=pd.to_numeric(s,errors='coerce')
print(numeric_s)

### tries to intepret each string as a number 
# Note:errrors='coerce'
# this is the key concept
# if converrsion is possible > converts manually
# if conversion###
```

```python
# convert numeric strings with commas to floats 
import pandas as pd
s=pd.Series(['1,000','2,500','3,750'])
# remove commas and convert to float
float_s=s.str.replace(',',"").astype(float)
# after replacing the commas converts strings to float type 
print(float_s)
```

```python
import pandas as pd

pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
# Sample data
data = {
    'School Code': ['s01', 's02', 's01', 's02', 's01'],
    'Student Name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
    'Score': [85, 90, 78, 92, 88],
    'index': [0, 1, 2, 3, 4]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Group by school code
student_groups = df.groupby('School Code')

# Iterate through groups
for name, group in student_groups:
    print("\nGroup for School Code:", name)
    print(group)

# Check type of GroupBy object
print("\nType of GroupBy object:")
print(type(student_groups))
```

```python
'''write a pandas program to merge two dataframes on a single coulmn'''
import pandas as pd

df1=pd.DataFrame({'ID':[1,2,3],
                  'Name':['John','Alice','Bob']})
df2=pd.DataFrame({'ID':[2,3,4],
                  'Age':[25,30,35]})
merged_df=pd.merge(df1,df2,on='ID',how='right')

#output result
print(merged_df)
```

```python
# Write a pandas program to draw a basic line plot with pandas and matplotlib.
# Create a sample DataFrame with 'Year' and 'Sales'
# Use plt.plot() to plot 'Year' vs 'Sales'
# Add axis labels and a title to the plot using plt.xlabel(), plt.ylabel(), and plt.title()
# Display the plot using plt.show()
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Sales': [100, 150, 200, 250, 300]
})

plt.plot(df['Year'], df['Sales'], marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Yearly Sales')
plt.show()
```
